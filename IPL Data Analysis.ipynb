```python
won_by_batting_first = data[data['win_by_runs']!=0]
```

```python
won_by_batting_first.head()
```

```python
plt.figure(figsize=(15,10))
plt.hist(won_by_batting_first['win_by_runs'])
plt.xlabel("Number of runs by which the team has won")
plt.ylabel("Number of instances")
plt.title("Distribution of winning runs")
```

```python
won_by_batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(12,8))
color = ['red','green','blue']
plt.bar(won_by_batting_first['winner'].value_counts()[0:3].keys(), won_by_batting_first['winner'].value_counts()[0:3],color=color)
```

```python
plt.figure(figsize=(8,8))
plt.pie(won_by_batting_first['winner'].value_counts(),labels = won_by_batting_first['winner'].value_counts().keys())
plt.title("Percentage of Wins by Batting First")
plt.show()
```

```python
won_by_batting_second = data[data['win_by_wickets']!=0]
won_by_batting_second.head()
```

```python
plt.figure(figsize=(15,10))
plt.hist(won_by_batting_second['win_by_wickets'],bins=30)
plt.xlabel("Number of wickets by which the team has won")
plt.ylabel("Number of instances")
plt.title("Distribution of winning wickets")
```

```python
won_by_batting_second['winner'].value_counts()
```

```python
plt.figure(figsize=(12,8))
color = ['red','green','blue']
plt.bar(won_by_batting_second['winner'].value_counts()[0:5].keys(), won_by_batting_second['winner'].value_counts()[0:5])
```

```python
plt.figure(figsize=(8,8))
plt.pie(won_by_batting_second['winner'].value_counts(),labels = won_by_batting_second['winner'].value_counts().keys())
plt.title("Percentage of Wins by Batting Second")
plt.show()
```
